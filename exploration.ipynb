```python
import seaborn as sns
sns.set_context("notebook")
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
import os

figsize=(7, 5)
```

### 14/02

```python
processed_14_02 = pd.read_csv("processed_data/Wednesday-14-02-2018_TrafficForML_CICFlowMeter.csv")
processed_14_02["Timestamp"] = pd.to_datetime(processed_14_02["Timestamp"], dayfirst=True, errors="coerce")
print("Dataset size:", processed_14_02.shape)
print("N missing timestamps:", processed_14_02["Timestamp"].isna().sum())
print("Class imbalance\n", processed_14_02["Label"].value_counts(), sep="")
plt.figure(figsize=figsize)
plt.title("14/02")
sns.histplot(data=processed_14_02.sample(10000), x="Timestamp", hue="Label");
plt.xticks(rotation=90);
plt.savefig("plots/14_02.png", dpi=300, bbox_inches="tight")
```

```python
def read_target():
    target = pd.read_csv("transformed_data/14_02/target.csv", header=None).sort_values(0)
    target = target[target[0] > 1e9]
    # target[0] = pd.to_datetime(target[0], unit="s")
    target = target.rename({0: "timestamp", 1: "target"}, axis=1)
    target["source"] = "target"
    return target

def read_files():
    files = os.listdir("transformed_data/14_02")
    files = [i.replace(".csv", "") for i in files]
    files.remove("target")
    files = pd.Series([int(i) for i in files]).sort_values()
    files = pd.DataFrame(files, columns=["timestamp"])
    files["source"] = "files"
    return files

    
target = read_target()
files = read_files()

print(target.shape, files.shape)
print(target.dtypes, files.dtypes)
```
