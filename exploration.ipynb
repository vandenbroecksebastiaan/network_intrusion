```python
import seaborn as sns
sns.set_context("notebook")
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
pd.set_option('display.float_format', lambda x: '%.3f' % x)

figsize=(7, 5)
```

### 14/02

```python
processed_14_02 = pd.read_csv("processed_data/Wednesday-14-02-2018_TrafficForML_CICFlowMeter.csv")
processed_14_02["Timestamp"] = pd.to_datetime(processed_14_02["Timestamp"], dayfirst=True, errors="coerce")
print("Dataset size:", processed_14_02.shape)
print("N missing timestamps:", processed_14_02["Timestamp"].isna().sum())
print("Class imbalance\n", processed_14_02["Label"].value_counts(), sep="")
plt.figure(figsize=figsize)
plt.title("14/02")
sns.histplot(data=processed_14_02.sample(10000), x="Timestamp", hue="Label");
plt.xticks(rotation=90);
plt.savefig("plots/14_02.png", dpi=300, bbox_inches="tight")
```

### Explore structure of packets

```python
target = pd.read_csv("transformed_data/14_02/target.csv", header=None).sort_values(0)
target = target[target[0] > 1e9]
cross_tab = pd.crosstab(index=target[0], columns=target[1])
cross_tab.to_csv("temp.csv")
plt.figure(figsize=(20, 10));
plt.subplot(3, 1, 1);
sns.scatterplot(y=cross_tab["Benign"], x=cross_tab.index, s=3)\
    .set_title("Number of packets per second stratified by origin");
plt.subplot(3, 1, 2);
sns.scatterplot(y=cross_tab["FTP-BruteForce"], x=cross_tab.index, s=3);
plt.subplot(3, 1, 3);
sns.scatterplot(y=cross_tab["SSH-Bruteforce"], x=cross_tab.index, s=3);
plt.savefig("plots/packets_per_second_origin.png", dpi=300, bbox_inches="tight")
```
