```python
import seaborn as sns
sns.set_context("notebook")
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_columns', 500)
pd.set_option('display.float_format', lambda x: '%.3f' % x)

figsize=(7, 5)
```

### 14/02

```python
processed_14_02 = pd.read_csv("processed_data/Wednesday-14-02-2018_TrafficForML_CICFlowMeter.csv")
processed_14_02["Timestamp"] = pd.to_datetime(processed_14_02["Timestamp"], dayfirst=True, errors="coerce")
print("Dataset size:", processed_14_02.shape)
print("N missing timestamps:", processed_14_02["Timestamp"].isna().sum())
print("Class imbalance\n", processed_14_02["Label"].value_counts(), sep="")
plt.figure(figsize=figsize)
plt.title("14/02")
sns.histplot(data=processed_14_02.sample(10000), x="Timestamp", hue="Label");
plt.xticks(rotation=90);
plt.savefig("plots/14_02.png", dpi=300, bbox_inches="tight")
```

### Explore structure of packets

```python
from multiprocessing import Pool
import os
from tqdm import tqdm

def generate_hexdump_scapy(file):
    from scapy.all import rdpcap, hexdump
    capture = rdpcap(f"original_data/Wednesday-14-02-2018/pcap/{file}")
    packet_time = [pd.to_datetime(float(i.time.real), unit="s") for i in capture]
    packet_hexdump = [hexdump(i, dump=True) for i in capture]
    packet_hexdump = [i.split("\n") for i in packet_hexdump]
    packet_hexdump = [[j.split("  ") for j in i] for i in packet_hexdump]
    packet_hexdump = [[j[1] for j in i] for i in packet_hexdump]
    packet_hexdump = [[j.replace(" ", "") for j in i] for i in packet_hexdump]
    packet_hexdump = ["".join(i) for i in packet_hexdump]
    return packet_hexdump, packet_time

def generate_hexdump_dpkt(file):
    print(file)
    import dpkt
    infile = dpkt.pcap.Reader(open(f"original_data/Wednesday-14-02-2018/pcap/{file}", "rb"))
    infile = [i for i in infile] # do not delete
    packet_time = [i[0] for i in infile]
    packet_hexdump = [i[1] for i in infile]

    # infile = [(i, dpkt.dpkt.hexdump(j, 1000)) for i, j in infile]
    with Pool(12) as pool:
        packet_hexdump = list(tqdm(pool.starmap(dpkt.dpkt.hexdump, [(i, 1000) for i in packet_hexdump])))

    # infile = [(i[0], i[1][9:1000]) for i in infile] 
    # infile = [(i[0], i[1].replace(" ", "")) for i in infile] 

    return packet_hexdump, packet_time

print(len(os.listdir("original_data/Wednesday-14-02-2018/pcap")))
files_02_03 = os.listdir("original_data/Wednesday-14-02-2018/pcap")[200:220]

# with Pool(12) as pool:
#     results = pool.map(generate_hexdump_dpkt, files_02_03)

results = [generate_hexdump_dpkt(i) for i in files_02_03]
    
# packet_hexdump = [i[0] for i in results]
# packet_hexdump = [i for j in packet_hexdump for i in j]
# packet_time = [i[1] for i in results]
# packet_time = [i for j in packet_time for i in j]
# 
```
